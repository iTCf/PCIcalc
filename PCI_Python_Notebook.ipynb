# Perturbational complexity index measure in TMS/EEG brain recordings.

In this Python Notebook a step by step calculation of the perturbational complexity index (PCI) [1] is presented. Herein we compare a wake to a sleep session of the same healthy subject and quantify the complexity of the corresponding EEG recording in terms of PCI.  
The notebook assumes that an inverse model (here the 3-spheres BERG method was used) has already been applied to the electrode level data and the corresponding sources (i.e. cortical currents) are therefore available. 

References:
[1] A Theoretically Based Index of Consciousness Independent of Sensory Processing and Behavior. A G Casali, O Gosseries, M Rosanova, M Boly, S Sarasso, K R Casali, S Casarotto, M Bruno, S Laureys, G Tononi, M Massimini. Sci Transl Med. 2013 Aug 14;5(198):198ra105. doi: 10.1126/scitranslmed.3006294.

```python
% matplotlib inline
import numpy as np
import pylab as plt
from scipy.io import loadmat   # to load mat files
import pci # PCI module downloaded from L.Barbosa's github and adapted for this Notebook
plt.rcParams['figure.figsize'] = (14,8)
```

```python
# parameters of the bootstrap statistics 
Nboot = 480  # number of bootstraps
alpha = 0.01 # alpha level of the statistical test
npoint_art = 4 # number of data points of the artifact
boot_precomputed = True #False # True = load previously generated, False = generate new bootstrap
pci_precomputed = True # True = load previously generated, False = computes the PCI
nind = 107  # number of data points to take (107 corresponds to 300 ms) - REPORT SAMPLING FREQUENCY !!!
```

```python
# file names to load - 
fnW = 'data/wake.mat'
fnS = 'data/sleep.mat'
```

Declare some procedures.

```python
def preprocess(data):
    ''' extract and preprocess relevant data 
    '''
    global nCH,nSRC,nTR
    AllTf = data['AllTf']                        # nSRC x nCH x nTR (num sources x num channels x num trials)
    Y = data['Y']                                # nCH x nTIME x nTR (num channels x num time steps x num trials)
    PCI = data['PCI'][0]
    times = data['times'][0]                     
    nSRC, nCH, nTR = AllTf.shape
    nTIME = Y.shape[1]  
    N0 = np.where( times < 0 )[0][-1]             # index of last negative time point
    J = np.tensordot(AllTf,Y,[[1,2],[0,2]])/nTR   # nSRC x nTIME (currents)
    basecorr = np.mean(J[:,:N0],axis=1)           # mean baseline of prestimulus activity       
    Norm = np.std(J[:,:N0],axis=1,ddof=1)         # standard deviation of baseline of prestimulus activity       
    J -= np.kron(np.ones((1,nTIME)),basecorr.reshape((nSRC,1)))  # subtract baseline
    NUM = np.kron(np.ones((1,N0)),basecorr.reshape((nSRC,1)))
    DEN = np.kron(np.ones((1,N0)),Norm.reshape((nSRC,1)))
    return AllTf, Y, PCI, times, N0, Y, J, NUM, DEN, Norm
```

The following bootstrap procedure allows to compute a threshold in order to determine the significant spatio-temporal cortical activations to the transcranial magnetic stimulation (TMS). A non-parametric bootstrap statistics (of parameter alpha) is computed  on the baseline and a threshold for each channel is obtained. 
The comparison with the threshold allows the determination of the binary spatio-temporal matrix of significant sources (matrix SS(x,t) of Fig.1 in Casali et al.2013).  

```python
def bootstrap(AllTf,Y,Norm,N0,NUM,DEN):
    ''' 
    non-parametric bootstrap procedure:
        - arguments:    
            AllTf, number of sources (nSRC) x number of channels (nCH) x number of trials (nTR) 
            Y, number of channels (nCH) x number of time steps (nTIME) x number of trials (nTR)
            Norm, N0, NUM, DEN
        - note: 
            all arguments come the "preprocess" procedure
    '''
    nSRC,_,nTR=AllTf.shape 
    nTIME=Y.shape[1]
    randontrialsT=np.random.randint(0,nTR,nTR)
    Bootstraps=np.zeros((Nboot,N0))     
    for per in xrange(Nboot):
        ET=np.zeros((nSRC,N0))              
        for j in xrange(nTR): 
            randonsampT=np.random.randint(0,N0,N0)    
            ET+=np.dot(AllTf[:,:,randontrialsT[j]],Y[:,randonsampT,randontrialsT[j]])        
        ET=(ET/nTR-NUM)/DEN # computes a Z-value 
        Bootstraps[per,:] = np.max(np.abs(ET),axis=0)       # maximum statistics in space
    # computes threshold for binarization depending on alpha value 
    Bootstraps=np.sort(np.reshape(Bootstraps,(Nboot*N0)))
    calpha=1-alpha 
    calpha_index=int(np.round(calpha*Nboot*N0))-1               
    TT=Norm*Bootstraps[calpha_index]                            # computes threshold based on alpha set before 
    Threshold=np.kron(np.ones((1,nTIME)),TT.reshape((nSRC,1)))  # set the same threshold for each time point 
    return Threshold
```

```python
# load data
dataW = loadmat(fnW)
dataS = loadmat(fnS)
```

```python
# get and preprocess data
AllTf_W, Y_W, PCI_W, times_W, N0_W, Y_W, J_W, NUM_W, DEN_W, Norm_W = preprocess(dataW)
AllTf_S, Y_S, PCI_S, times_S, N0_S, Y_S, J_S, NUM_S, DEN_S, Norm_S = preprocess(dataS)
```

In the field of TMS/EEG recordings butterfly plots are commonly used to get some indications on the complexity of the responses. The butterfly plots are obtained by averaging the activity the time course of each channel over trials. 
In the following we report the butterfly plot from a healthy subject for a sleep and a wake session. 

```python
 # Butterfly plot 
Yavg_W = Y_W.mean(axis=2)    # average each channel over trials 
Yavg_S = Y_S.mean(axis=2)    # average each channel over trials 
fig, axs = plt.subplots(1,2)
for i in xrange(nCH): 
    axs[0].plot(times_S,Yavg_S[i,:],'b-')
    axs[1].plot(times_W,Yavg_W[i,:],'r-')
axs[0].set_title('Sleep',fontsize=18)    
axs[1].set_title('Wake',fontsize=18)    
axs[0].set_xlabel('time (ms)',fontsize=16)
axs[1].set_xlabel('time (ms)',fontsize=16)
plt.suptitle('Butterlfly plot',fontsize=20)
```

A visual inspection of the butterfly plot shows that the response to TMS is much slower during sleep than during wake. This already provides some hints on the "expected" complexity that we will quantify with the PCI measurement.

```python
# determine threshold of significant spatio-temporal events
if boot_precomputed:
    print "Load precomputed bootstraps from a previous run." 
    Threshold_W=np.load("data/Threshold_W.npy")
    Threshold_S=np.load("data/Threshold_S.npy")   
else:
    print "Computes the bootstraps." 
    print ">>> This is going to take some minutes!"
    Threshold_W=bootstrap(AllTf_W,Y_W,Norm_W,N0_W,NUM_W,DEN_W)
    Threshold_S=bootstrap(AllTf_S,Y_S,Norm_S,N0_S,NUM_S,DEN_S)
    np.save("data/Threshold_W.npy",Threshold_W)
    np.save("data/Threshold_S.npy",Threshold_S)
```

```python
print J_W.shape
plt.plot(times_W,J_W[0,:],'k-')


# determines sources matrices 
binJ_W=np.array(np.abs(J_W)>Threshold_W,dtype=int) 
binJ_S=np.array(np.abs(J_S)>Threshold_S,dtype=int) 
```

In order to compute the minimal complexity of the spatio-temporal patterns we rank all channels by their activity. As stated in Casali et al. 2013 [1] this ordering yields a proxy of the minimum complexity of the significant sources matrix.

```python
# rank the activity matrix - use mergesort that yields same results of Matlab
Irank_W=np.argsort(np.sum(binJ_W,axis=1),kind='mergesort') 
binJrank_W=np.copy(binJ_W)
binJrank_W=binJ_W[Irank_W,:]
Irank_S=np.argsort(np.sum(binJ_S,axis=1),kind='mergesort') 
binJrank_S=np.copy(binJ_S)
binJrank_S=binJ_S[Irank_S,:]
```

We report the plots of the significant sources (x coordinate time, y coordinate space). Based on these plots we conclude that based on the boostrap statistics any pre-stimulus activity is not signficant (no black dots at t < 0) while the  response reflects the activation of deterministic and complex spatio-temporal patterns.

```python
# plot ranked binJ
c_W,r_W=np.where(binJrank_W[::-1,:]) # ::-1 reverse ordering for plot
c_S,r_S=np.where(binJrank_S[::-1,:])
# 
fig, axs = plt.subplots(1,2)
axs[0].plot(times_S[r_S],c_S,'ko')
axs[1].plot(times_W[r_W],c_W,'ko')
axs[0].set_xlabel('time (ms)',fontsize=16)
axs[1].set_xlabel('time (ms)',fontsize=16)
axs[0].set_xlim(xmin=-200,xmax=times_S[-1]*1.05) # 1.05*times_S[0]
axs[1].set_xlim(xmin=-200,xmax=times_W[-1]*1.05) # 1.05*times_W[0]
axs[0].set_ylim(ymax=nSRC)
axs[1].set_ylim(ymax=nSRC)
axs[0].set_ylabel('sources ranked by activity',fontsize=16)
fig.suptitle('significant source matrices',fontsize=20)
axs[0].set_title('Sleep',fontsize=18)
axs[1].set_title('Wake',fontsize=18)
```

As a final step we compute the perturbational complexity index and plot its time course. The end point of the PCI curve corresponds to the complexity of the spatio-temporal patterns. 

```python
# computes PCI
ind_binJ_S=N0_S+npoint_art+np.arange(nind)
ind_binJ_W=N0_W+npoint_art+np.arange(nind)
if pci_precomputed: 
    ct_S=np.load("data/pci_S.npy")
    ct_W=np.load("data/pci_W.npy")
    print "Loaded precomputed bootstraps from a previous run." 
else:        
    binJ_S=binJrank_S[:,ind_binJ_S]
    binJ_W=binJrank_W[:,ind_binJ_W]
    ct_S=pci.lz_complexity_2D(binJ_S)
    ct_W=pci.lz_complexity_2D(binJ_W)
    norm_S=pci.pci_norm_factor(binJ_S)
    norm_W=pci.pci_norm_factor(binJ_W)
    ct_S=ct_S/norm_S
    ct_W=ct_W/norm_W
```

```python
# plot PCI 
yMAX=max(ct_S.max(),ct_W.max())*1.05
fig, axs = plt.subplots(1,2)
axs[0].plot(times_S[ind_binJ_S],ct_S,'k-') #,label='Python')
#axs[0].plot(times_S[ind_binJ_S],PCI_S,'r-',label='Matlab')
axs[1].plot(times_W[ind_binJ_W],ct_W,'k-')
#**********axs[1].plot(times_W[ind_binJ_W],PCI_W,'r-')
axs[0].set_xlabel('time (ms)',fontsize=16)
axs[1].set_xlabel('time (ms)',fontsize=16)
axs[0].set_ylabel('PCI',fontsize=16)
axs[0].set_ylim(ymax=yMAX)
axs[1].set_ylim(ymax=yMAX)
fig.suptitle('Perturbation Complexity Index (PCI)',fontsize=20)
axs[0].set_title('Sleep',fontsize=18)
axs[1].set_title('Wake',fontsize=18)
xS,yS=times_S[ind_binJ_S][-1],ct_S[-1]
xW,yW=times_W[ind_binJ_W][-1],ct_W[-1]
axs[0].plot(xS,yS,'ko',markersize=8)
axs[1].plot(xW,yW,'ko',markersize=8)
axs[0].annotate('PCI=%2.2g'%yS,xy=(xS,yS),xytext=(xS-100,yS-0.1),arrowprops=dict(facecolor='black', shrink=0.05),fontsize=14)
axs[1].annotate('PCI=%2.2g'%yW,xy=(xW,yW),xytext=(xW-100,yW-0.1),arrowprops=dict(facecolor='black', shrink=0.05),fontsize=14)
```

These plots show that the complexity of the spatio-temporal patterns is significantly lower during the sleep session respect to the wake session.
